```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inp = np.load('Input_data_ML.npy')
out = np.load('Output_data_ML.npy')
```

```python
n=np.shape(inp)[0]
print(n)
```

```python
inp=inp[1000:n:10,:]
out=out[:,1000:n:10]
```

```python
plt.plot(inp[1000])
print(out[:,1000])
```

```python
Cov=np.transpose(inp)@inp
```

```python
eig,W=np.linalg.eig(Cov)
np.shape(W)
```

```python
y=np.abs(inp@np.transpose(W[:2,:]))
np.shape(y)
```

```python
for i in range(15):
    cm = plt.cm.get_cmap('RdYlBu')
    sc = plt.scatter(y[:,0], y[:,1], c=out[0], s=1, cmap=cm)
    plt.colorbar(sc)
    plt.show()
```
