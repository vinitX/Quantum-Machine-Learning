```python
import qiskit
from qiskit import QuantumRegister, QuantumCircuit, execute
from qiskit.compiler import transpile
import numpy as np
from qiskit.quantum_info.operators import Operator
```

```python
q=3
ckt = qiskit.QuantumCircuit(q)
A=np.random.rand(2**q,2**q)
U,S,V=np.linalg.svd(A)

ckt.unitary(U,list(range(q)))

transpile(ckt,basis_gates=['u3','cx']).draw(output='mpl')
```

```python
q=3
q=3
ckt = qiskit.QuantumCircuit(q+1)
A=np.random.rand(2**(q+1),2**q)
U,S,V=np.linalg.svd(A)

ckt.isometry(U,list(range(q)),[q])
ckt.draw()
```
