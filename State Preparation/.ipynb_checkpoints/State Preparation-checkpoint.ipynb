```python
from qiskit import QuantumRegister, QuantumCircuit, execute
from qiskit.quantum_info.operators import Operator
from qiskit.visualization import plot_histogram, plot_state_city
from qiskit import Aer
from qiskit.compiler import transpile
import numpy as np
from math import sin, cos, atan2
import matplotlib.pyplot as plt
from numpy.linalg import norm,svd
```

```python
def state_prep(ckt,A,q):
    if q==1:
        ckt.u3(2*atan2(A[1],A[0]),0,0,0)
        return ckt
    
    n=q-q//2 #n>=m
    m=q//2
    A=np.reshape(A,(2**m,2**n))
    A=np.transpose(A)
    
    U,S,V=np.linalg.svd(A)
    ckt=state_prep(ckt,S,q//2)
    
    for i in range(m):
        ckt.cx(i,i+n)
    if q%2==1:
        ckt.cx(q//2,q-1)
    ckt.barrier()
    
    ckt.unitary(Operator(U), list(range(n)))
    ckt.unitary(Operator(np.transpose(V)), list(range(n,q)))
    ckt.barrier()
    print(q,"Q   done")
    return ckt
```

```python
q=2
n=q-q//2 #n>=m
m=q//2

A=np.random.rand(2**q)
A=A/np.linalg.norm(A)

ckt=QuantumCircuit(q,q)
ckt=state_prep(ckt,A,q)
ckt.draw(output='mpl')
```

```python
ckt.data
```

```python
backend = Aer.get_backend('statevector_simulator')
job=execute(ckt,backend)
result = job.result()
outputstate = result.get_statevector(ckt, decimals=6)

#print(np.real(outputstate))
#print(np.round(A,3),'\n')
```
